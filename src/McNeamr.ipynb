# McNemar's test for model comparison

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
```

```python
a, b = 32, 1
chi_stat = (a-b)**2/(a+b)

df = 1
p_value = 1 - chi2.cdf(chi_stat, df)

print('Chi-squared statistics for McNeamr\'s test:', chi_stat)
print('P-value for McNeamr\'s test:', p_value)
```

```python
observed = chi_stat

x = np.linspace(0, 30, 500)
pdf = chi2.pdf(x, df)
observed_pdf = chi2.pdf(observed, df)

plt.figure(figsize=(5, 3))
plt.plot(x, pdf, label=f"Chi-squared Distribution (df={df})", color="blue")
plt.scatter(observed, observed_pdf, color="red", label=f"Observed = {observed}", marker='*', s=100)

plt.title("Chi-squared Distribution with Observed Value")
plt.xlabel("Chi-squared statistics")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)

plt.show()
```
